- dropna() - This function allows you to drop all or some rows that have missing values
- fillna() - This function allows you to replace the rows that have missing values with the value what you want pass

```python
len(df)# gives the number of rows in the data
```

```python
new_df=df.dropna(axis=0,how='any')# row wise checks for null value and drops the row if data is not there
```

## Visualizing data
- An interesting way to diaplay DataFrame
- With matplotlib

```python
import matplotlib.pyplot as plt
a1=[1,2,3,4]
a2=[5,6,7,8]
plt.figure(figsize=(6,6))
plt.plot(a1,a2,color='blue')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
marks=[12,60,76,43,21,100]
divisions=['A','B','C','D','E','F']
avg_marks=[12,2,23,3,45,33]
plt.bar(divisions,marks,color="blue",width=0.70)
```

```python
plt.bar(avg_marks,marks,color="green",width=0.70)
```

```python
plt.barh(divisions,marks,color="#1412cd")
```

```python
plt.barh(avg_marks,divisions,color="violet")
```

```python
lang=['python','java','c','c++']
rating=[75,70,55,65]
plt.pie(rating,labels=lang,shadow=True,startangle=90,autopct='%1.1f%%')#%1.1f%%
plt.legend(loc='best')
plt.show()
```
