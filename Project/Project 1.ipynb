## Problem Statement
- Plot a bar graph no. of matches won by each team in the year 2011

## Description of the problem
- data frame is resticted to year 2011 and is grouped by match winners and the number of times each team won is counted using
- country id
- Then the data frame is converted into lists of indexes and values
- then a bar graph is drawn 

```python
# q2
b=df[df['Season_Year']==2011].groupby('match_winner')['Country_id'].count()
plt.figure(figsize=(25,10))
q=b.index.tolist()
w=b.values.tolist()
plt.bar(q,w,color='blue',width=0.5)
```
