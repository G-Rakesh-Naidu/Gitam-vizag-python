```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('F:/data science/Data/AppleStore.csv')
df
```

## Problem Statement
- 1.Calculate the average rating of Gaming and Non-Gaming apps 

## Description of the problem
- The data frame frame is divided into gaming and non gaming apps
- From those data we calculated the means of user ratings

```python
a=df[df['prime_genre']=='Games']
b=df[df['prime_genre']!='Games']
print("AVG USER RATING OF GAMING APPS IS",a['user_rating'].mean())
print("AVG USER RATING OF NON GAMING APPS IS",b['user_rating'].mean())
```

## Problem Statement
- 2.Draw Pie plot based on labels as below Price category and values as count of each category apps 
- labels:
    - Price =0 - Free apps
    - Price <20 - Affordable apps
    - Price<50 - Expensive apps
    - Price >50 - Very expensive

## Problem Description
- apps are categorized into free,affordable,expensive,very expensive
    - they are taken as labels for our pie chart
- list li is appended with the count of these apps
- Then a pie chart is drawn with the list li and apps names that are categorized

```python
apps=['Free apps','Affordable apps','Expensive apps','Very expensive']
li=[]
li.append(df[df['price']==0]['id'].count())
li.append(df[(df['price']>0) & (df['price']<20)]['id'].count())
li.append(df[(df['price']>20) & (df['price']<50)]['id'].count())
li.append(df[df['price']>50]['id'].count())
plt.figure(figsize=(8,8))
plt.pie(li,labels=apps,startangle=90,autopct='%1.1f%%')
plt.legend()
```
